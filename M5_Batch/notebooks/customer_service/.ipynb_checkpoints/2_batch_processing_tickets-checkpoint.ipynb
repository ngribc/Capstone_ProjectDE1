```python
# =========================
# 1. LOAD DATA
# =========================

df = spark.read.option("header", True) \
    .csv("data/raw/customer_support_tickets.csv")

df.printSchema()
df.show(5)
```

```python
#🔹 Limpieza
from pyspark.sql.functions import col

# Normalizar nombres
df_clean = df.select(
    [col(c).alias(c.lower().replace(" ", "_")) for c in df.columns]
)

df_clean.show(5)
```

```python
#🔹 Tipado (MUY IMPORTANTE PRO)
from pyspark.sql.functions import col

# Normalizar nombres
df_clean = df.select(
    [col(c).alias(c.lower().replace(" ", "_")) for c in df.columns]
)

df_clean.show(5)
```
