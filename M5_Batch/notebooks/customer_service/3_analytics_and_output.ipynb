```python
#🔹 Queries tipo DWH
df = spark.read.parquet("data/lake/tickets")

df.createOrReplaceTempView("tickets")
```

```python
# 🔹 Query 1 (Distribución categórica)
spark.sql("""
SELECT status, COUNT(*) as total
FROM tickets
GROUP BY status
ORDER BY total DESC
""").show()
```

```python
#🔹 Query 2 (Temporal)
spark.sql("""
SELECT DATE(created_at) as date, COUNT(*) as total
FROM tickets
GROUP BY date
ORDER BY date
""").show()
```

```python
#🔹 Guardar resultados (para BI)
result = spark.sql("""
SELECT status, COUNT(*) as total
FROM tickets
GROUP BY status
""")

result.write.mode("overwrite").parquet("data/output/status_summary")
```
